```python
#3)bargraph
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
new_data = sns.load_dataset("fmri")


# Create a bar plot with a specific order for the x-axis
sns.barplot(x="timepoint", y="signal", hue="region", data=new_data, hue_order=["parietal", "frontal"],ci=30,saturation=10,errcolor="r",errwidth=3,dodge=True)


#order gives  sequenced series
#hue order means to have main(hue column) components in sequenced order
#ci=/n_boot black colour line in bar will change according to given number
#orient used to make  bar horizontal or vertical
#use pallete to change the colour
#saturation used to blur
#err is dandi on top errcolor to change color err width to change width
#dodge = falseused to make single plot
#alpha=0.2,0.3,1,2 used to change the shade of color of bars


plt.show()
```

```python
#ii>
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# creating a specific array using linesace()
tru=np.linspace(1,10,15)#making array of range lyinhg between 1 to 10 and including 15 elements
x=tru.reshape(5,3)#reshaping the array
arr1=np.array([["a1","a2","a3"],["b1","b2","b3"],["c1","c2","c3"],["d1","d2","d3"],["e1","e2","e3"]])
annot1={"color":"y"}#creating for use in heatmap use annot_kws for calling

#calling heatmap function
sns.heatmap(x,annot=arr1,fmt="s",annot_kws=annot1)#anoot meands givin / showing values in graph and fmt=s(because its a string type of data)
```

```python
#8)violin plot
#i>
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
v_d=sns.load_dataset("tips")
print("this is tips dataset:\n\n",v_d.head(100))
c=sns.violinplot(x="day",y="total_bill",data=v_d)
v=sns.violinplot(x="day",y="total_bill",data=v_d,hue="time",linewidth=2,color="r",palette="Dark2",split=True,alpha=1)# using hue this  means comparing with something
plt.show()
```

```python
#ii>
#heree we can use inner parameter in violin plot
#'box': Displays a miniature box plot inside each violin. This shows the median, interquartile range (IQR), and potential outliers, providing additional insights into the distribution.
##'quartile': Displays the quartiles of the data (25th, 50th, and 75th percentiles) as lines inside each violin.
#'point': Displays all the individual data points within the violin.
#'stick': Draws lines for each individual data point inside the violin, providing a denser view.
#None: Disables the display of any internal data points, showing only the outer shape of the violin plot.
#8)violin plot
#i>points
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
v_d=sns.load_dataset("tips")
c=sns.violinplot(x="day",y="total_bill",data=v_d)
v=sns.violinplot(x="day",y="total_bill",data=v_d,hue="time",linewidth=2,
                 color="r",palette="Dark2",split=True,alpha=1,inner="point")
# using hue this  means comparing with something
plt.show()
```
